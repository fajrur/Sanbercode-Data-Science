<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]

def kali_list(obj_list, obj_penambah):
    a = obj_list
    b = obj_penambah
    return(a + b)
    
    
print(kali_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
