Jawab: 

```python
# Apa perbedaan Series dengan list dan dictionary
## Series merupakan objek satu dimensi yang berisi 1 data yang seragam serta memiliki index dan key. 
## Sedangkan list merupakan kumpulan data yang bisa berbeda-beda tipe datanya dan hanya memiliki index.
## Sedangkan dictionary merupakan kumpulan data yang mempunyai keys dan value.

# Apa perbedaan dataframe dan series
## Dataframe merupakan objek dua dimensi dan merupakan kumpulan dari series.
## Series merupakan obejek dengan single list.

# Buatlah suatu data series
country = pd.Series(["Japan", "Thailand", "Indonesia", "New Zealand", "Switzerland"])
country
```

Di soal no 2 kamu telah membaca data tentang diamonds. pada soal ini cobalah suatu metode dari pandas untuk mengetahui beberapa data pertama dan beberapa data terakhir. kemudian sebutkan nilai dari:
- baris pertama dari column price

<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!

Deskriptif statistik merupakan suatu metode statistik yang berhubungan dengan pengumpulan atau penyajian data sampai memberi informasi yang berguna.

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
print(diamonds.describe())

# Berapa nilai rata-rata dari column price
print(diamonds["price"].mean())

# Berapa nilai standar deviasi dari column depth
print(diamonds["depth"].std())

# Berapa nilai maximum dari column carat
print(diamonds["carat"].max())
```
