```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
group = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
group
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
import numpy as np
grouptrans = group.transpose()

plt.figure(figsize = (18,6))

x = np.arange(0, 16, 4)
width = 0.3
colors = ['red', 'blue','purple', 'gray', 'gold', 'green', 'pink', 'red', 'blue', 'purple', 'gray', 'gold']

for i, j in enumerate(grouptrans.columns):
  plt.bar(x + (i * width), grouptrans[j], width, label = j, color = colors[i])
 
plt.style.use('ggplot')
plt.xticks(x + 1.6, labels = grouptrans.index)
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.title('Mean Sales Video Games By Genre')
plt.legend(loc=(0.95,0.75), frameon=True)
plt.show()
```

Jawab:
Dengan menggunakan barplot, kita dapat mengetahui perbandingan banyaknya data antar kategori. Jika barplotnya semakin tinggi berarti datanya di suatu kategori banyak dan jika semakin pendek berarti datanya di suatu kategori sedikit.

Dibandingkan dengan region lain, rata-rata penjualan video games tertinggi adalah North America. Untuk penjualan per genrenya, di North America lebih banyak yang membeli game dengan genre Platform. Di Europe lebih banyak yang membeli game dengan genre Shooter. Di Japan lebih banyak yang membeli game dengan genre Role-Playing. Sedangkan di wilayah lain lebih banyak yang membeli game dengan genre Shooter.

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
plt.figure(figsize=(18, 6))

x = grouptrans.index
width = 0.7
y_bottom = 0

colors = ['orange', 'blue', 'purple', 'gray', 'gold', 'green', 'pink', 'orange', 'blue', 'purple', 'gray', 'gold']

for i,j in enumerate(grouptrans.columns):
  plt.bar(x, grouptrans[j], width, label = j, color = colors[i], bottom = y_bottom)
  y_bottom += grouptrans[j]
 
plt.style.use('ggplot')
plt.legend(loc=(0.95,0.75), frameon=True)
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.title('Mean Sales Video Games By Genre')
plt.show()
```

Jawab:
Dengan Stack Barplot, kita dapat membandingkan banyaknya data tiap kategori dengan tinggi plot masing-masing. Semakin tinggi plotnya, berarti data tersebut semakin banyak. Sedangkan semakin rendah plotnya, berarti data tersebut semakin sedikit.  

Dibandingkan dengan region lain, rata-rata penjualan video games tertinggi adalah North America.
