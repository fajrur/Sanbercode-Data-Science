```python
df_number2 = pd.read_csv('austin_weather.csv', na_values = ['na', '-'])
df_number2.dtypes
```

```python
df_number2.info()
```

```python
df_nohandling = df_number2.sample(1000)
```

```python
df_nohandling.isnull().sum()
```

```python
plt.figure(figsize = (12, 8))
plt.scatter(x = df_nohandling['HumidityAvgPercent'], y = df_nohandling['DewPointAvgF'], c = df_nohandling['TempAvgF'], s = df_nohandling['WindAvgMPH'] * 20, cmap = 'coolwarm')
plt.xlabel('HumidityAvg (%)')
plt.ylabel('DewPointAvg (F)')
plt.title('Austin Weather')
plt.show()
```

```python
df_number2.fillna(method = 'ffill', inplace = True)
```

```python
df_withhandling = df_number2.sample(1000)
```

```python
df_withhandling.isnull().sum()
```

```python
plt.figure(figsize = (12, 8))
plt.scatter(x = df_withhandling['HumidityAvgPercent'], y = df_withhandling['DewPointAvgF'], c = df_withhandling['TempAvgF'], s = df_withhandling['WindAvgMPH'] * 20, cmap = 'coolwarm')
plt.xlabel('HumidityAvg (%)')
plt.ylabel('DewPointAvg (F)')
plt.title('Austin Weather')
plt.show()
```

Analisa : Pada visualiasi dengan handling missing value, ada perbedaan beberapa data yang persebarannya acak sehingga beberapa data tidak sesuai (karena value NaN diisi dengan nilai pada data sebelumnya). Semakin tinggi nilai DewPointAvg dan HumidityAvg maka Temperatur juga akan semakin tinggi, begitu juga sebaliknya.

<h2>Soal 3. Visualisasi Data with Handling outliers </h2>

Pada tugas kali ini, kita akan melakukan handling outliers. Outliers ditemukan pada kolom 'House_Price' dengan menggunakan boxplot dari library seaborn seperti gambar berikut: 


![boxplot1](https://drive.google.com/uc?id=1y2jRwf8zRPXQtSBkmjJLcqUm4l6DR7d9)


Dengan menggunakan Interquartile range, kita bisa temukan nilai data outliers dan meremove data outlier tersebut. 

*   Tentukan nilai batas atas dan batas bawah dari Interquartile range tersebut
*   Remove data outliers tersebut menggunakan batas atas dan batas bawah
*   Visualisasikan hasil data yang sudah d remove outliers seperti expected berikut:







**EXPECTED OUTPUT:**

Batas bawah & batas atas:

871625.0 10976625.0

Visualisasi boxplot:

![boxplot2](https://drive.google.com/uc?id=1qZc-lEa5FUzSQ_cyt4_GJQ0TZ8c6WjHX)





```python
Q1 = df2['House_Price'].quantile(0.25)
Q3 = df2['House_Price'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - IQR * 1.5
upper = Q3 + IQR * 1.5

non_outlier = df2[(df2.House_Price < upper) & (df2.House_Price > lower)]
```

```python
print('batas bawah & batas atas:')
print(lower, upper)
print('Visualisasi boxplot:')

import seaborn as sns
plt.figure(figsize = (12, 8))
sns.boxplot(x = 'House_Price', data = non_outlier)
plt.show()
```
